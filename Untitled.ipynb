 Question 2: Use Webscraping to Extract Tesla Revenue Data

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

# Set the URL and headers to look like a browser
url = "https://www.macrotrends.net/stocks/charts/TSLA/tesla/revenue"
headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/114.0.0.0 Safari/537.36"
}

# Send GET request
response = requests.get(url, headers=headers)

# Parse the HTML content
soup = BeautifulSoup(response.content, "html.parser")

# Find all tables
tables = soup.find_all("table")

# The revenue table is usually the second table on the page
for i, table in enumerate(tables):
    print(f"Table {i}:\n{table.text[:100]}")  # preview table text

# Assuming the correct one is tables[1]
revenue_table = tables[1]

# Convert to DataFrame
tesla_revenue = pd.read_html(str(revenue_table))[0]

# Clean the data
tesla_revenue.columns = ['Date', 'Revenue']
tesla_revenue['Revenue'] = tesla_revenue['Revenue'].replace({'\$': '', ',': ''}, regex=True)
tesla_revenue.dropna(inplace=True)
tesla_revenue = tesla_revenue[tesla_revenue['Revenue'] != ""]

# Display the last 5 rows
tesla_revenue.tail()
```

Question 4: Use Webscraping to Extract GME Revenue Data

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
from io import StringIO  # Import StringIO to wrap html string

# URL for GME revenue data
url = "https://www.macrotrends.net/stocks/charts/GME/gamestop/revenue"

# Fetch webpage content
html_data = requests.get(url).text

# Parse HTML
soup = BeautifulSoup(html_data, "html.parser")

# Find all tables on the page
tables = soup.find_all("table")

# Find the correct revenue table (contains "GameStop Quarterly Revenue" text)
for table in tables:
    if "GameStop Quarterly Revenue" in table.text:
        revenue_table = table
        break

# Wrap the HTML string with StringIO before passing to read_html
gme_revenue = pd.read_html(StringIO(str(revenue_table)))[0]

# Clean the dataframe
gme_revenue.columns = ["Date", "Revenue"]
gme_revenue["Revenue"] = gme_revenue["Revenue"].replace("[\$,]", "", regex=True)
gme_revenue.dropna(inplace=True)
gme_revenue = gme_revenue[gme_revenue["Revenue"] != ""]

# Display last 5 rows
print(gme_revenue.tail())
```

```python
import matplotlib.pyplot as plt

def make_graph(df, title):
    plt.figure(figsize=(12,6))
    plt.plot(df['Date'], df['Close'], label='Close Price')
    plt.xlabel('Date')
    plt.ylabel('Closing Price (USD)')
    plt.title(title)
    plt.xticks(rotation=45)
    plt.legend()
    plt.grid(True)
    plt.show()

# Example usage with your Tesla stock data DataFrame (tesla_data)
# Make sure the 'Date' column is in datetime format
tesla_data['Date'] = pd.to_datetime(tesla_data['Date'])

make_graph(tesla_data, "Tesla Stock Closing Prices Over Time")
```
